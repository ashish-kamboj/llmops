1. Create an Ollama account at [ollama.com](https://ollama.com)
2. Generate an API key from [ollama.com/settings/keys](https://ollama.com/settings/keys)
3. Set the API key in the `.env` file in this folder

### Installation

````python
# Setup: Install required packages and import libraries
import sys
import subprocess

packages = ['ollama', 'python-dotenv', 'requests']

for package in packages:
    try:
        __import__(package.replace('-', '_'))
    except ImportError:
        print(f"Installing {package}...")
        subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", package])

print("âœ“ All packages installed successfully!")
````

````python
# Import all required libraries
import os
from pathlib import Path
from dotenv import load_dotenv
import ollama
from typing import Dict, Any

# Load environment variables from .env file
# Use Path.cwd() since __file__ is not defined in Jupyter notebooks
env_path = Path.cwd() / '.env'
if env_path.exists():
    load_dotenv(env_path)
else:
    # Try to find .env in the ollama-cloud directory
    load_dotenv(dotenv_path='d:\\ollama-n8n\\ollama-cloud\\.env')

# Get API key and configuration
OLLAMA_API_KEY = os.getenv('OLLAMA_API_KEY')
OLLAMA_ENDPOINT = os.getenv('OLLAMA_API_ENDPOINT', 'https://ollama.com')
DEFAULT_MODEL = os.getenv('DEFAULT_MODEL', 'gpt-oss:120b')

# Verify API key is set
if not OLLAMA_API_KEY or OLLAMA_API_KEY == 'your_api_key_here':
    print("WARNING: OLLAMA_API_KEY not set in .env file!")
    print("Please set your API key in the .env file: https://ollama.com/settings/keys")
else:
    print(f"âœ“ Ollama API configured")
    print(f"  Endpoint: {OLLAMA_ENDPOINT}")
    print(f"  Default Model: {DEFAULT_MODEL}")

# Initialize Ollama client for cloud API
client = ollama.Client(
    host=OLLAMA_ENDPOINT,
    headers={'Authorization': f'Bearer {OLLAMA_API_KEY}'}
)
````

## Web Search Capability - Access Real-time Information

Web search allows models to search the internet and access current information, reducing hallucinations and enabling the model to provide up-to-date answers.

**Features:**
- Real-time web search
- Document retrieval
- Web content fetching
- Building AI agents that can research topics

````python
def test_web_search(query: str = "What is Ollama?", max_results: int = 5) -> Dict[str, Any]:
    """
    Test web search capability - get real-time web search results.
    
    Args:
        query: Search query
        max_results: Maximum number of results to return
    
    Returns:
        Web search results
    """
    print(f"   Performing web search...")
    print(f"   Query: '{query}'")
    print(f"   Max Results: {max_results}\n")
    
    try:
        # Use the configured client for web search
        results = client.web_search(query, max_results=max_results)
        
        print(f"âœ“ Found {len(results.results)} results:\n")
        
        for i, result in enumerate(results.results, 1):
            print(f"{i}. {result.title}")
            print(f"   URL: {result.url}")
            print(f"   Content: {result.content[:200]}...")
            print()
        
        return {'query': query, 'results_count': len(results.results), 'results': results.results}
    
    except Exception as e:
        print(f"   Web search error: {str(e)}")
        print("   Note: Ensure your API key has web search permissions")
        return {'error': str(e)}

def test_web_fetch(url: str = "https://ollama.com") -> Dict[str, Any]:
    """
    Test web fetch capability - fetch and parse a web page.
    
    Args:
        url: URL to fetch
    
    Returns:
        Page content, title, and links
    """
    print(f"   Fetching web page...")
    print(f"   URL: {url}\n")
    
    try:
        # Use the configured client for web fetch
        result = client.web_fetch(url)
        
        print(f"âœ“ Page fetched successfully\n")
        print(f"Title: {result.title}")
        print(f"\nContent (first 300 chars):\n{result.content[:300]}...")
        print(f"\nLinks found: {len(result.links)}")
        for link in result.links[:5]:
            print(f"  - {link}")
        if len(result.links) > 5:
            print(f"  ... and {len(result.links) - 5} more")
        
        return {'url': url, 'title': result.title, 'links_count': len(result.links)}
    
    except Exception as e:
        print(f"  Web fetch error: {str(e)}")
        return {'error': str(e)}

# Test web search and fetch
print("=" * 60)
print("WEB SEARCH CAPABILITY")
print("=" * 60)
search_result = test_web_search("Ollama cloud models", max_results=3)

print("\n" + "=" * 60)
print("WEB FETCH CAPABILITY")
print("=" * 60 + "\n")
fetch_result = test_web_fetch("https://ollama.com")
````
