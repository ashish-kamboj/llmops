# Data Pipeline (HTML -> Text -> Embeddings -> Clickstream)
Run the steps in order:
1. Generate synthetic Confluence-like HTML pages
2. Extract clean text from HTML
3. Create Gemini embeddings and store them in Postgres
4. Seed users and clickstream data for personalization


```python
import subprocess
import sys
from pathlib import Path

root = Path.cwd()

def run_script(path: Path) -> None:
    result = subprocess.run([sys.executable, str(path)], capture_output=True, text=True)
    if result.returncode != 0:
        raise RuntimeError(result.stderr or "Script failed")
    if result.stdout:
        print(result.stdout)

run_script(root / "src" / "pipelines" / "generate_pages.py")
print("Generated HTML pages")
```
